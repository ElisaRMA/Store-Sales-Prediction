### Created Functions

```python
def cramer_v(x,y):
    # criar tabela de contingencia - crosstab 
    # criando para duas aleatoriamente, primeiro. depois, quando aplica dentro da função, deixa isso generico 
    # e testa varias outras 
    contingency = pd.crosstab(x, y).as_matrix()

    # passar tabela pro chi e extrair o resultado do teste
    chi2 = chi2_contingency(contingency)[0]

    # # n é soma de todos os valores, é o mesmo que o nnumero de linhas do datset original
    n = contingency.sum()

    r,k = contingency.shape

    # # correcao do bias

    nom_corr = max(0, (chi2/n) - ((k-1)*(r-1)/(n-1)))

    k_corr = k - (k-1)**2 / n-1  

    r_corr = r - (r-1)**2/ n-1 

    V = np.sqrt((nom_corr) / min(k_corr-1, r_corr-1))
    
    return V



def ml_error (model_name, y, yhat):
    mae = mean_absolute_error(y, yhat)
    mape = mean_absolute_percentage_error(y, yhat)
    rmse = np.sqrt(mean_squared_error(y, yhat))
    
    return pd.DataFrame({'Model Name': model_name,
                         'MAE': mae,
                         'MAPE': mape,
                         'RMSE': rmse}, index=[0])

warnings.filterwarnings('ignore')


def cross_validation(x_training,kfold,model_name, model):
  results = []
  for k in reversed(range(1,kfold+1)):
    
    x_training['date'] = pd.to_datetime(x_training['date'])
    validation_start_date = x_training['date'].max() - datetime.timedelta(days=k*6*7)
    validation_end_date = x_training['date'].max()- datetime.timedelta(days=(k-1)*6*7)

    training = x_training[x_training['date'] < validation_start_date]
    validation = x_training[(x_training['date'] >= validation_start_date) & (x_training['date'] <= validation_end_date)]

    # training dataset
    xtraining = training.drop(['date', 'sales'], axis=1)
    ytraining = training['sales']

    # validation
    xvalidation = validation.drop(['date','sales'], axis=1)
    yvalidation = validation['sales']

    # model
    
    pipeline = Pipeline([('preprocessing', processing_pipe),
                         ('model', model)])
    
    m = pipeline.fit(xtraining, ytraining)

    # prediction
    yhat = m.predict(xvalidation)

    # performance
    m_result = ml_error(model_name, np.expm1(yvalidation), np.expm1(yhat))
    results.append(m_result)
    modelling_result = pd.concat(results)

    mae = modelling_result['MAE'].mean()
    mape = modelling_result['MAPE'].mean()
    rmse = modelling_result['RMSE'].mean()

    mae_std= modelling_result['MAE'].std()
    mape_std= modelling_result['MAPE'].std()
    rmse_std= modelling_result['RMSE'].std()


  return pd.DataFrame({'Model Name': model_name,
                      'MAE CV': [mae,mae_std], 
                        'MAPE CV': [mape,mape_std],
                        'RMSE CV': [rmse, rmse_std]},index=['mean','std'])
```

## Fill NA's

```python
data1['competition_distance'] = data1['competition_distance'].apply(lambda x: 200000.0 if math.isnan(x) else x)


data1['competition_open_since_month'] = data1.apply(lambda x: x['date'].month 
                                                            if math.isnan(x['competition_open_since_month']) 
                                                            else x['competition_open_since_month'], axis=1)


data1['competition_open_since_year'] = data1.apply(lambda x: x['date'].year 
                                                           if math.isnan(x['competition_open_since_year']) 
                                                           else x['competition_open_since_year'], axis=1)

data1['promo2_since_week'] = data1.apply(lambda x: x['date'].week 
                                                 if math.isnan(x['promo2_since_week']) 
                                                 else x['promo2_since_week'], axis=1)


data1['promo2_since_year'] = data1.apply(lambda x: x['date'].year 
                                                 if math.isnan(x['promo2_since_year']) 
                                                 else x['promo2_since_year'], axis=1)

month_map = {1: 'Jan', 2: 'Feb', 3: 'Mar', 
             4:'Apr', 5:'May', 6:'Jun', 
             7:'Jul', 8:'Aug', 9:'Sept', 
             10:'Oct', 11:'Nov', 12:'Dec'}

data1['promo_interval'].fillna(0, inplace=True)
data1['month_map'] = data1['date'].dt.month.map(month_map)

data1['is_promo'] = data1[['promo_interval', 'month_map']].apply(lambda x: 0 
                                                                         if x['promo_interval'] == 0 
                                                                         else 1 
                                                                         if x['month_map'] in x['promo_interval'].split(',') 
                                                                         else 0, axis=1)
```

## Feature Engineering

```python
data2['year'] = data2['date'].dt.year

data2['month'] = data2['date'].dt.month

data2['day'] = data2['date'].dt.day

data2['week_of_year'] = data2['date'].dt.weekofyear

data2['year_week'] = data2['date'].dt.strftime('%Y-%W')



data2['competition_since'] = data2.apply(lambda x: datetime.datetime(year =x['competition_open_since_year'] , 
                                                                     month =x['competition_open_since_month'], 
                                                                     day=1),axis=1)

data2['competition_time_month'] = ((data2['date'] - data2['competition_since'].values.astype('datetime64[ns]'))/30).apply(lambda x: x.days).astype(int)


data2['promo_since'] = data2['promo2_since_year'].astype(str) + '-' + data2['promo2_since_week'].astype(str)

data2['promo_since'] = data2['promo_since'].apply(lambda x: datetime.datetime.strptime(x + '-1', '%Y-%W-%w') - datetime.timedelta(days=7))

data2['promo_time_week'] = ((data2['date'] - data2['promo_since'])/7).apply(lambda x: x.days).astype(int)

data2['assortment'] = data2['assortment'].apply(lambda x: 'basic' if x == 'a' 
                                                else 'extra' if x == 'b' 
                                                else 'extended')

data2['state_holiday'] = data2['state_holiday'].apply(lambda x: 'public_holiday' if x == 'a' 
                                                else 'easter_holiday' if x == 'b' 
                                                else 'christmas' if x == 'c' else 'regular_day')
```

- CRAMER V 

https://en.wikipedia.org/wiki/Cram%C3%A9r%27s_V

v = np.sqrt( (chi2/n) / min(k-1, r-1) )

n - observations
r - rows
k - columns 

```python
# combinações
v1 = cramer_v(cat['state_holiday'], cat['state_holiday'])
v2 = cramer_v(cat['state_holiday'], cat['store_type'])
v3 = cramer_v(cat['state_holiday'], cat['assortment'])

v4 = cramer_v(cat['store_type'], cat['state_holiday'])
v5 = cramer_v(cat['store_type'], cat['store_type'])
v6 = cramer_v(cat['store_type'], cat['assortment'])

v7 = cramer_v(cat['assortment'], cat['state_holiday'])
v8 = cramer_v(cat['assortment'], cat['store_type'])
v9 = cramer_v(cat['assortment'], cat['assortment'])

# dataframe de combinações
cramer_vs = {'state_holiday': [v1, v2, v3],
             'store_type': [v4, v5, v6],
             'assortment': [v7, v8, v9]}    

cat_correlation = pd.DataFrame(cramer_vs)    

cat_correlation.index = ['state_holiday', 'store_type','assortment']
cat_correlation
```

```python
# Scaling

rs = RobustScaler()
mms = MinMaxScaler()

# competition distance
X_train['competition_distance'] = rs.fit_transform(X_train[['competition_distance']].values)

# # competition time month
X_train['competition_time_month'] = rs.fit_transform(X_train[['competition_time_month']].values)

# year
X_train['year'] = mms.fit_transform(X_train[['year']].values)

# promo time week
X_train['promo_time_week'] = mms.fit_transform(X_train[['promo_time_week']].values)
```

```python
# Encoding

# state holiday - One Hot
X_train = pd.get_dummies(X_train, prefix = ['state_holiday'], columns = ['state_holiday'])

# store_type - nao tem ordem 
le = LabelEncoder()
X_train['store_type'] = le.fit_transform(X_train['store_type'])

# assortment - ordinal
assortment_dict = {'basic':1,
                  'extra': 2,
                  'extended': 3}

X_train['assortment'] = X_train['assortment'].map(assortment_dict)
```

# Machine Learning Modeling

## Pipeline

```python
# Pipeline 

rs = RobustScaler()
mms = MinMaxScaler()
ohe = OneHotEncoder()
oe = OrdinalEncoder(categories=[['basic', 'extra', 'extended']], dtype='int') 

processing_pipe = ColumnTransformer([
                                 ('robust-scaler', rs, ['competition_distance', 'competition_time_month']),
                                 ('minmax', mms, ['promo_time_week']),
                                 ('ohe', ohe,['store_type']),
                                 ('ordinal_assortment', oe , ['assortment'])], 
                               remainder = 'passthrough')
```

## Average Model

```python
# no need for pipeline

# making a copy of the x to use in each model
aux1 = x_test.copy()
aux1['sales'] = y_test.copy()

# taking the mean, the first model, the column predictions has the means
aux2 = aux1[['store','sales']].groupby('store').mean().reset_index().rename(columns={'sales':'predictions'})

# uniting the 
aux1 = pd.merge(aux1,  aux2, how='left', on='store')
yhat_baseline = aux1['predictions']

# performance
baseline_result= ml_error('Average Model', np.expm1(y_test), np.expm1(yhat_baseline))
baseline_result
```

```python
# sum of predictions
data91 = data9[['store','predictions']].groupby('store').sum().reset_index()

# MAE e MAPE - não fala mae e mape, só retorna o erro - bruto e em porcentagem. 
data9_aux1 = data9[['store', 'sales', 'predictions']].groupby('store').apply(lambda x: mean_absolute_error(x['sales'], x['predictions'])).reset_index().rename(columns = {0:'MAE'})
data9_aux2 = data9[['store', 'sales', 'predictions']].groupby('store').apply(lambda x: mean_absolute_percentage_error(x['sales'], x['predictions'])).reset_index().rename(columns = {0:'MAPE'})

# merge
data9_aux3 = pd.merge(data9_aux1, data9_aux2, how='inner', on='store')
data92 = pd.merge(data91, data9_aux3, how='inner', on='store')

# pior cenário é predição - erro ou seja, vendendo menos
# melhor cenário é prediçao  + erro, ou seja, vende mais! 
data92['worst_scenario'] = data92['predictions'] - data92['MAE']
data92['best_scenario'] = data92['predictions'] + data92['MAE'] 

#order columns 
data92 = data92[['store', 'predictions','worst_scenario', 'best_scenario', 'MAE', 'MAPE']]
```

## API Handler

```python
# file Handler.py 

from flask import Flask, request, Response
from rossmann.Rossmann import Rossmann
import pickle
import pandas as pd
import os 

# loading model - pipeline 
model = pickle.load(open('model/pipeline.pkl', 'rb'))

# initialize app
app = Flask(__name__)

@app.route('/rossmann/predict', methods=['POST'])
def rossmann_predict():
    test_json = request.get_json()

    if test_json:
        if isinstance(test_json, dict):
            test_raw = pd.DataFrame(test_json, index=[0])
        else: 
            test_raw = pd.DataFrame(test_json, columns = test_json[0].keys())

        #instantiate rossmann
        initial_pipeline = Rossmann()

        # data cleaning
        data1 = initial_pipeline.data_cleaning(test_raw)

        # feature eng
        data2 = initial_pipeline.feature_engineering(data1)

        # data prep
        data3 = initial_pipeline.data_prep(data2)

        # prediction
        data_response = initial_pipeline.get_prediction(model, test_raw, data3)

        return data_response

    else:
        return Response('{}', status=200, mimetype='application/json')


if __name__ == '__main__':
    port =  os.environ.get('PORT', 5000)
    app.run(host='0.0.0.0', port=port)
```
